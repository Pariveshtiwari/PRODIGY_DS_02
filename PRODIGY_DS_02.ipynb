# Data Cleaning

```python
fifa.shape
```

```python
fifa.info
```

```python
fifa.describe()
```

```python
fifa.isnull().sum()
```

```python
# Calculating the mean of the 'GKPositioning' column, ignoring nan values

mean_GK_positioning = fifa['GKPositioning'].mean()
```

```python
# Replace nan values in the 'GKPositioning' column with the mean value
mean_GK_positioning=fifa['GKPositioning'].fillna(mean_GK_positioning,inplace=True)
```

```python
fifa['GKPositioning'].isna().sum() #verify that there is no nan values left in the 'GKPositionning column'
```

```python
fifa.GKPositioning.unique()
```

### removing null values from the whole dataset 

```python
#Check for null values in the dataset
print("Number of nan values in each column before replacement:\n",fifa.isna().sum())
```

```python
#identify numeric columns
numeric_cols=fifa.select_dtypes(include='number').columns
```

```python
#Replace nan values in numeric columns with the mean of the respective columns
fifa[numeric_cols] = fifa[numeric_cols].apply(lambda x:x.fillna(x.mean()))
```

```python
print("Number of nan values in the each column after replacement:\n",fifa.isna().sum())
```
